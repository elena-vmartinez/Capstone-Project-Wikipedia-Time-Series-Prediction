```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/train_1.csv')
print(df.shape)
```

```python
df.head()
```

```python
df['agent_type'] = df['Page'].apply(lambda x: x.split('_')[-1])
df_humanish = df[~df['Page'].str.endswith('_spider')].copy()
print("After removing spiders:", df_humanish.shape) # subset exluding pure bot traffic to avoid spikes and focus on real-user popularity
```

```python
subset_pages = df_humanish['Page'].unique()[:10] # first 10 pages
df_subset = df_humanish[df_humanish['Page'].isin(subset_pages)].copy()
print(df_subset.shape)
print(df_subset.isnull().sum().sum())
```

```python
df_long = df_subset.melt(id_vars=['Page', 'agent_type'], var_name='Date', value_name='Views')
df_long['Date'] = pd.to_datetime(df_long['Date'], format='%Y-%m-%d', errors='coerce') # use date as single column
df_long.head()
```

```python
df_long = df_long.sort_values(['Page', 'Date'])
df_long['Views'] = df_long.groupby('Page')['Views'].ffill() # using forward-fill to fill missing visits with previous days' data
```

```python
print(df_long.isnull().sum())
df_long.head()
```

```python
df_long = df_long.dropna(subset=['Views'])
print(df_long.isnull().sum())
df_long.head()
```

```python
import matplotlib.pyplot as plt

sample_pages = df_long['Page'].unique()
for page in sample_pages:
    page_data = df_long[df_long['Page'] == page]
    plt.figure(figsize=(12, 4))
    plt.plot(page_data['Date'], page_data['Views'])
    plt.title(f"Page Views Over Time: {page}")
    plt.xlabel("Date")
    plt.ylabel("Views")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

Working with singular page https://fr.wikipedia.org/wiki/@ . It shows a stable, low-noise traffic pattern over time. The page itself doesn't have much content but its data is ideal for evaluating time series models.

```python
import numpy as np

chosen_page = '@_fr.wikipedia.org_desktop_all-agents'
my_page = df_long[df_long['Page'] == chosen_page].copy()
my_page = my_page.sort_values('Date')

my_page['Lag1'] = my_page['Views'].shift(1) # adding 1-day lag feat
my_page = my_page.dropna().copy()

test_days = 30 # predicting last 30 days

train = my_page[:-test_days].copy()
test = my_page[-test_days:].copy()

y_train = train['Views'].values
y_test = test['Views'].values

train['DayIndex'] = np.arange(len(train))
test['DayIndex'] = np.arange(len(train), len(train) + len(test)) # adding weekday feat for seasonality

train['Weekday'] = train['Date'].dt.weekday
test['Weekday'] = test['Date'].dt.weekday

X_train = train[['DayIndex', 'Lag1', 'Weekday']]
X_test = test[['DayIndex', 'Lag1', 'Weekday']]

print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
```

```python
from sklearn.metrics import mean_absolute_error

def evaluate_mae(model_name, y_true, y_pred, mae_dict):
    mae = mean_absolute_error(y_true, y_pred)
    mae_dict[model_name] = mae
    print("MAE -", model_name, ":", round(mae, 2))
    return mae

mae_dict = {}
```

```python
baseline_pred = np.repeat(y_train[-1], len(y_test)) # using last value in training data to predict every day in the test set
evaluate_mae("Baseline", y_test, baseline_pred, mae_dict)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
lr_pred = lr.predict(X_test)

evaluate_mae("Linear Regression", y_test, lr_pred, mae_dict)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 4))
plt.plot(test['Date'], y_test, label='Actual', linewidth=2)
plt.plot(test['Date'], lr_pred, label='Linear Regression', linestyle='--')
plt.plot(test['Date'], baseline_pred, label='Baseline', linestyle=':')
plt.title("Linear Regression vs Baseline")
plt.xlabel("Date")
plt.ylabel("Views")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import xgboost as xgb

xgb_model = xgb.XGBRegressor(objective='reg:squarederror', n_estimators=100, random_state=42)
xgb_model.fit(X_train, y_train)
xgb_pred = xgb_model.predict(X_test)

evaluate_mae("XGBoost", y_test, xgb_pred, mae_dict)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 4))
plt.plot(test['Date'], y_test, label='Actual', linewidth=2)
plt.plot(test['Date'], baseline_pred, label='Baseline', linestyle=':', color='green')
plt.plot(test['Date'], lr_pred, label='Linear Regression', linestyle='--', color='orange')
plt.plot(test['Date'], xgb_pred, label='XGBoost', linestyle='-.', color='purple')

plt.title("Model Comparison: Baseline vs Linear Regression vs XGBoost")
plt.xlabel("Date")
plt.ylabel("Views")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam

model = Sequential()
model.add(Dense(256, activation='relu', input_shape=(X_train.shape[1],)))
model.add(Dropout(0.3))

model.add(Dense(128, activation='relu'))
model.add(Dropout(0.3))

model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1))

model.compile(optimizer=Adam(learning_rate=0.001), loss='mae')

from tensorflow.keras.callbacks import EarlyStopping

early_stop = EarlyStopping(
    monitor='val_loss',
    patience=10,
    restore_best_weights=True
)

history = model.fit(
    X_train, y_train,
    epochs=200,  # more time to learn, but will stop early if needed
    batch_size=16,
    validation_split=0.1,
    callbacks=[early_stop],
    verbose=1
)

nn_pred = model.predict(X_test).flatten()

evaluate_mae("Neural Net (Dense)", y_test, nn_pred, mae_dict)
```

```python
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss (MAE)')
plt.title('Training vs Validation Loss')
plt.legend()
plt.grid(True)
plt.show()
```

## Scaling

```python
from tensorflow.keras import Input

date_cols = df_humanish.columns.drop(['Page', 'agent_type'], errors='ignore') # only columns with daily views included

df_subset = df_humanish.sample(10, random_state=42).set_index('Page') # sampling 10 more pages instead of 1 for scaling
mae_scores = {}

# Data prep for every series for correct input + output
def create_supervised_data(series, lookback=30): # Creates supervised data, X for past 30 days of data and Y for actual # of visits the next day
    X, y = [], []
    for i in range(len(series) - lookback):
        X.append(series[i:i+lookback])
        y.append(series[i+lookback])
    return np.array(X), np.array(y)
```

```python
for page in df_subset.index:
    print(f"Training model for: {page}")

    try:
        ts = df_humanish[df_humanish['Page'] == page][date_cols].values.flatten()
        ts = pd.Series(ts).interpolate().bfill().ffill().values

        if len(ts) < 60: # Need 60 days of recent data total-30 for history, 30 to predict
            print(f"Skipping {page}: not enough data.")
            continue

        X, y = create_supervised_data(ts, lookback=30) # Calls data prep function
        X_train, y_train = X[:-30], y[:-30]
        X_test, y_test = X[-30:], y[-30:]

        model = Sequential()
        model.add(Input(shape=(X_train.shape[1],)))
        model.add(Dense(256, activation='relu'))
        model.add(Dropout(0.3))
        model.add(Dense(128, activation='relu'))
        model.add(Dropout(0.3))
        model.add(Dense(64, activation='relu'))
        model.add(Dense(32, activation='relu'))
        model.add(Dense(1))
        model.compile(optimizer=Adam(learning_rate=0.001), loss='mae')

        early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

        history = model.fit(
            X_train, y_train,
            epochs=200,
            batch_size=16,
            validation_split=0.1,
            callbacks=[early_stop],
            verbose=0
        )

        y_pred = model.predict(X_test).flatten()
        mae = mean_absolute_error(y_test, y_pred)
        mae_scores[page] = mae

    except Exception as e:
        print(f"Skipping {page} due to error: {e}")
```

```python
mae_table = pd.DataFrame(mae_scores.items(), columns=["Page", "MAE"]).sort_values(by="MAE")
mae_table
```

### Scaling the Model to Multiple Time Series

To demonstrate the scalability of my deep learning forecasting model, I applied it to 10 randomly sampled Wikipedia pages. Each pageâ€™s time series was transformed into supervised learning format using a 30-day lookback window to predict the next day. The model was re-trained independently for each page.

The MAE varied significantly across pages. This variation reflects differences in traffic volume and volatility. Pages with consistent traffic patterns had lower MAEs, while trending or seasonal pages showed much higher errors due to their complex behavior.

In a real-world deployment, further improvements could include scaling normalization, outlier filtering, or separate model tuning per traffic tier. This exercise demonstrates that the pipeline generalizes across different types of time series and is scalable to larger datasets.


### Trade-offs When Scaling the Forecasting Model

**Computation time**: Training a deep learning model 10 times increases training time significantly. This would be even more impactful with hundreds or thousands of time series.

**Model reusability vs. accuracy**: I chose to retrain a new model per page for better accuracy. A shared model could be faster but less precise due to the diversity in traffic patterns.

**Varying performance**: Some pages had consistent traffic and low MAE, while others had large fluctuations or high traffic, leading to higher errors. This highlights the need for tailored modeling or normalization when scaling broadly.

While I didn't train across the entire dataset of 145,000+ pages due to time and resource constraints, I created a pipeline that can be scaled further with additional time and resources.
